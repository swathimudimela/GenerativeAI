```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
text_to_classify = "I am stuck on what to read in genAI"
```

```python
classification_result = classifier(text_to_classify)
print(f"Predicted Sentiment is : {classification_result}")
```

```python
print(f"The text is classified as {classification_result[0]['label']}")
```

```python
# Lets try the named entity recognition
ner_pipeline = pipeline('ner')
sample_text = " I like to do online shopping on amazon"

ner_results = ner_pipeline(sample_text)
```

```python
print(f"Named entities Found : ")

for entity in ner_results:
    print(f"Entity : {entity['word']} , label : {entity['entity'] }, score : {entity['score']}")
```

```python
# table question Answering

import pandas as pd
```

```python
data = {"Actors" : ["A" , "B" , "C"] , 
        "Number of Scenes" : ["87" , "53" , "67"]
        }

# convert this dictionary to dataframe ie table
table = pd.DataFrame.from_dict(data)

question = "How many scenes does actor B have"

tqa = pipeline(task = "table-question-answering" , model = "google/tapas-large-finetuned-wtq")
print(tqa(table = table, query = question)['cells'][0])
```

```python
# normal Question Answer
qa_pipeline = pipeline("question-answering")

context = " I love doing online shopping when i want to do shopping"
question = "where do I like to shop"

qa_result = qa_pipeline(question = question, context = context)

print(f"Answer : {qa_result['answer']}")
```

```python
# question ansering qith Zero- shot prompting
zero_shot_classifier = pipeline('zero-shot-classification')

text_to_classify = " The new samrtphone has impressive camera features"

candidate_labels = ["Technology", "Fashion", "Sports"]

classification_result = zero_shot_classifier(text_to_classify, candidate_labels)

print(f"Predicted_label is {classification_result['labels'][0]} , with confidence {classification_result['scores'][0]}")
```

```python
# translation english-> french

translator = pipeline("translation" , model = "google-t5/t5-small")

text_to_translate = " I am trying to translate this line"

translation_result = translator(text_to_translate)

print(f" Translated Text : {translation_result[0]['translation_text']}")
```
